# Data Aggregation:

```python
#a. Find max value in all columns using groupby function on'continent'column
df_groupby=covid.groupby("continent").agg({"location":"max","date":"max","total_cases":"max","new_cases":"max","new_cases_smoothed":"max","total_deaths":"max","new_deaths":"max","new_deaths_smoothed":"max","gdp_per_capita":"max","extreme_poverty":"max","cardiovasc_death_rate":"max","diabetes_prevalence":"max","female_smokers":"max","male_smokers":"max","handwashing_facilities":"max","hospital_beds_per_thousand":"max","life_expectancy":"max","human_development_index":"max"})
```

```python
df_groupby
```

```python
#a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' column to 'total_cases
df_groupby["Deaths ratio"]=df_groupby["total_cases"]/df_groupby["total_deaths"]
```

# 9. Data Visualization

```python
import seaborn as sns

#a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.

sns.distplot(df_groupby["gdp_per_capita"])
```

```python
#b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
sns.jointplot(data=df_groupby,x="total_cases",y="gdp_per_capita",kind="scatter")
```

```python
#c. Plot Pairplot on df_groupby dataset

sns.pairplot(data=df_groupby)
```

```python
#d. Plot a bar plot of 'continent' column with 'total_cases' .
import matplotlib.pyplot as plt
plt.figure(figsize=(5,10))
sns.catplot(data=df_groupby1,x="continent",y="total_cases",kind="bar")
plt.xlabel("continent")
plt.ylabel ("total_cases")
plt.title("Insights")
```
